```python
import pandas as pd
import pymongo
```

```python
from pymongo import MongoClient

client = MongoClient('mongodb://localhost:27017/')
database = client['exercices']
collection = database['youtube']
```

```python
df_youtube = pd.read_csv("./data/USvideos.csv")
df_youtube.head(2)
```

## Question 0

### Netoyer les données

```python
df_youtube2=df_youtube.sort_values("likes",ascending=False)
```

```python
df_youtube2.head()
```

```python
df_youtube.isnull().sum()
```

```python
df_youtube.info()
```

### Importer les données

```python
df_json=pd.read_json("./data/US_category_id.json")
df_json["category_id"]=df_json["items"].apply(lambda x: int(x["id"]))
df_y=df_youtube2.merge(df_json, on='category_id')
```

```python
df_y.head(3)
```

```python
collection.insert_many(df_y.to_dict(orient='records'))
```

## Question 5
- 5) Compter le nombre moyen de vues en fonction de la catégorie.
- 5）按类别计算平均浏览次数。

```python
list(collection.aggregate([{"$group" : {"_id" : "$category_id", "moyen de vues " : {"$avg" : "$views"}}}]))
```
